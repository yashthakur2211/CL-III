```python
import pandas as pd

# Sample data: Date, Temperature, Location
data = [
    ("2023-06-01", 35.5, "New York"),
    ("2023-07-15", 38.0, "New York"),
    ("2022-12-10", -5.0, "Chicago"),
    ("2022-08-01", 30.0, "Chicago"),
    ("2023-11-05", 10.0, "New York"),
    ("2021-05-21", 25.5, "Houston"),
]

df = pd.DataFrame(data, columns=["Date", "Temperature", "Location"])
df
```

```python
# Map Step
df['Year'] = df['Date'].str[:4]  # Extract year
mapped_data = df[['Year', 'Temperature']]
mapped_data
```

```python
# Shuffle & Sort Step
grouped = mapped_data.groupby('Year')
```

```python
# Reduce Step
result = grouped.agg(Min_Temperature=('Temperature', 'min'),
                     Max_Temperature=('Temperature', 'max'))
result
```
